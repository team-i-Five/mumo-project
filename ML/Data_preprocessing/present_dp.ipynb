### 텍스트 전처리

```python
# 형태소 분리를 위한 Konlpy 객체 초기화
twitter = Okt()
```

```python
#텍스트를 전처리 하는 함수
def preprocessingText(synopsis):
    #전처리한 결과를 저장 할 리스트
    stems = []
    #text 를 단어로 나누고 품사를 붙여서 리턴
    #stem=True : 단어의 기본형으로 변환  ex) 같습니다,같았습니다 -> 같다.  
    tagged_review = twitter.pos(synopsis,stem=True)

    #tagged에서 하나를 꺼내서 word,pos에 대입
    #단어는 word, 품사는 pos에 대입
    for word,pos in tagged_review:
        #len(word) : 단어의 글자수가 2글자 이상이면
        #품사가 Noun(명사) 또는 Adjective(형용사) 일경우에만 word(단어) 를 stems에 추가
        if pos=="Noun" or pos=='Adjective':
            stems.append(word)

    #리스트인 stems를 문자열로 만들어서 ( "".join(stems) ) 리턴       
    return " ".join(stems)
```

```python
# 데이터프레임의 'synopsis' 열에 NaN 값이 있을 때 빈 문자열로 대체 및 전처리
present_data['synopsis_clear'] = present_data['synopsis'].fillna('').apply(preprocessingText)
```

### 벡터를 문자열로 저장

```python
#행의 수를 row_num에 저장
row_num = len(present_data)
row_num
```

```python
string_array = []  # 문자열을 저장할 빈 리스트

for index in range(row_num):
    # 뮤지컬 번호 조회
    NUM = present_data.loc[index,"musical_id"]
    print("NUM=",NUM)
    # 뮤지컬 제목 조회
    TITLE = present_data.loc[index,"title"]
    print("TITLE=",TITLE)
    # 줄거리 조회
    LINE = present_data.loc[index,"tokenized_data"]
    print("LINE=",LINE)
    doc2vec = None
    count = 0
    for word in LINE: #줄거리에서 단어 1개씩 리턴
        print("word=",word, end="\t")
        if word in word2vec_words: #단어가 WordVector에 포함된 단어 이면
            count += 1 #개수 1증가
            if doc2vec is None: #doc2vect이 None(초기값 맨처음 실행 첫번째 단어) 이면
                doc2vec = model.wv[word] #단어의 WordVector 값 대입
            else: #첫번째 단어 아님
                doc2vec = doc2vec + model.wv[word] #단어의 WordVector 값 추가

    if  doc2vec is not None:
        # 단어 벡터를 모두 더한 벡터의 값을 단어의 개수로 나눠줌
        doc2vec = doc2vec / count
     # 문자열로 변환해서 string_array 리스트에 추가
    string_array.append(doc2vec.tostring())

# string_array에 저장된 문자열을 데이터프레임에 추가
present_data["synopsis_vector"] = string_array
```

### date 분리

```python
# 날짜 형식을 "YYYY/MM/DD"에서 "YYYY-MM-DD" 형식의 문자열로 변환
# 문자열을 "~"를 기준으로 나누고, 각각 시작 날짜와 끝 날짜로 분리합니다.
date_split = present_data['date'].str.split(' ~ ')

# 시작 날짜를 가져와서 "YYYY/MM/DD"에서 "YYYY-MM-DD" 형식의 문자열로 변환합니다.
present_data['start_date'] = pd.to_datetime(date_split.str[0], format="%Y/%m/%d", errors='coerce')

# 끝 날짜를 가져와서 "YYYY/MM/DD"에서 "YYYY-MM-DD" 형식의 문자열로 변환합니다.
present_data['end_date'] = pd.to_datetime(date_split.str[1], format="%Y/%m/%d", errors='coerce')
```

### tag

```python
selected_genre = ["웃음", "친구", "아이", "사람", "슬픔", "사랑", "마법", "공포", "역사", "오페라", "예술"]
similar_words_by_genre = {}

for genre in selected_genre:
    # 모델 어휘에 있는 경우에만 유사한 단어 찾기
    if genre in model.wv.key_to_index:
        similar_words = model.wv.most_similar(genre, topn=1000)
        similar_words_by_genre[genre] = similar_words
    else:
        print(f"'{genre}' is not present in the vocabulary.")

# 결과 출력 또는 다른 작업 수행
print(similar_words_by_genre)
```

```python
# 'tag' 열 초기화
for i in range(1, 4):
    present_data[f'tag{i}'] = None
present_data
```

```python
# 장르 할당
for index, row in present_data.iterrows():
    synopsis = row['synopsis']
    assigned_genres = set()  # 중복을 허용하지 않는 세트로 변경

    for genre in selected_genre:
        if not pd.isna(synopsis) and genre in synopsis:
            assigned_genres.add(genre)  # 세트에 장르 추가

    if len(assigned_genres) < 3 and not pd.isna(synopsis):
        max_similar_genres = set()
        max_similarity = 0

        for genre, similar_words in similar_words_by_genre.items():
            if genre not in assigned_genres:  # 이미 태그로 지정된 장르가 아닌 경우에만 처리
                for word, similarity in similar_words:
                    if word in synopsis and similarity > max_similarity:
                        max_similar_genres = {genre}
                        max_similarity = similarity

        assigned_genres.update(max_similar_genres)

    assigned_genres = list(assigned_genres)[:3]  # 최대 3개까지만 저장

    for i, genre in enumerate(assigned_genres):
        present_data.at[index, f'tag{i + 1}'] = genre

    # tag2와 tag3에 NaN 값 설정
    if len(assigned_genres) < 2:
        present_data.at[index, 'tag2'] = 'NaN'
    if len(assigned_genres) < 3:
        present_data.at[index, 'tag3'] = 'NaN'

# 결과 확인
print(present_data[['title', 'synopsis', 'tag1', 'tag2', 'tag3']])
```
